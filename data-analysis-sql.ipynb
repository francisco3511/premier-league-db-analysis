## 2021/2022 Premier League Analysis

Perform a simple analysis, using SQL, of 2021/2022 season of the Premier League. 



```python
import sqlalchemy as sa

engine = sa.create_engine('postgresql://postgres:@localhost:5432/football_stats')
```

```python
%%sql 

SELECT * 
FROM pl_clubs 
```

```python
%%sql 

SELECT *
FROM pl_players
LIMIT 20
```

Find out the top 10 goal scorers and assisters:

```python
%%sql
SELECT player, team, goals
FROM pl_players
WHERE goals IS NOT NULL
ORDER BY goals DESC
LIMIT 10;
```

```python
%%sql
SELECT player, team, assists
FROM pl_players
WHERE assists IS NOT NULL
ORDER BY assists DESC
LIMIT 10;
```

```python
%%sql
SELECT team, SUM(goals) AS total_goals
FROM pl_players 
GROUP BY team
ORDER BY total_goals DESC;
```

```python
%%sql
SELECT player, team, goals
FROM pl_players as pl
WHERE goals = (
    SELECT MAX(goals)
    FROM pl_players
    WHERE team = pl.team
)
```

```python
%%sql
SELECT *
FROM pl_players pl INNER JOIN pl_clubs cl ON pl.team = cl.team
LIMIT 50;
```
